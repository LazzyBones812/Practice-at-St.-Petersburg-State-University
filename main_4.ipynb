```python
import osmnx as ox

places = ["Peterhof", "Strelna", "Saint Petersburg"]
G = ox.graph_from_place(places, network_type='drive')
ox.plot_graph(G)
```

```python
import networkx as nx

G1 = ox.graph_from_place("Peterhof", network_type='drive')
G2 = ox.graph_from_place("Strelna", network_type='drive')
G3 = ox.graph_from_place("Saint Petersburg", network_type='drive')
G4 = ox.graph_from_place("Moscow", network_type='drive')
graphs = [G1, G4]
full_graph = nx.compose_all(graphs)

ox.plot_graph(full_graph)
```

```python
graphs = [G3, G4]
full_graph_1 = nx.compose_all(graphs)

ox.plot_graph(full_graph_1)
```

```python
graphs = [G1, G3]
full_graph = nx.compose_all(graphs)

ox.plot_graph(full_graph)
```

```python
gdf_nodes, gdf_relationships = ox.graph_to_gdfs(full_graph_1)
gdf_nodes
```

```python
north, south, east, west = gdf_nodes[['y']].max(), gdf_nodes[['y']].min(), gdf_nodes[['x']].min(), gdf_nodes[['x']].max()
G = ox.graph_from_bbox(bbox = (north, south, east, west), network_type='drive')
ox.plot_graph(G)
```
